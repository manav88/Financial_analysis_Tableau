```sql
SELECT histgrJan24.Industry_Name, 
        histgrJan24.Number_of_Firms,
        CAGR_in_Net_Income_Last_5_years,
        CAGR_in_Revenues_Last_5_years,
        Expected_growth_in_Revenues_Next_2_years,
        Expected_Growth_in_EPS_Next_5_years,

        Gross_Margin,
        Net_Margin,
        EBITDA_Sales,

        Current_PE,
        PEG_Ratio,

        ROE_unadjusted
            
FROM histgrJan24
JOIN marginJan24 ON histgrJan24.Industry_Name = marginJan24.Industry_Name
JOIN pedataJan24 ON histgrJan24.Industry_Name = pedataJan24.Industry_Name
JOIN roeJan24 ON histgrJan24.Industry_Name = roeJan24.Industry_Name;
```
